working with JSON files--create and write on json file

```python
d1 = {'fname': 'Steve', 'lname': 'Jobs'}
d2 = {'fname': 'Steve', 'lname': 'Jobs', 'age': 21}

import json
wp = open(os.path.join('C:\\DSML_REV\\', 'test_json.json'), 'w+')
for x in range(5):
    if x == 0:
        wp.write(json.dumps(d1))
    elif x == 1:
        wp.write(json.dumps(d2))
    else:
        wp.write(json.dumps({'age':24, 'salary':10000}))
    wp.write('\n')
wp.close()  
wp = open(os.path.join('C:\\DSML_REV\\', 'test_json.json'), 'r')
for line in wp:
    print(wp.read())
```

```python
# ---working with JSON files--read json file .get and rec['key'] have the same functionality

rp = open(os.path.join('C:\\DSML_REV\\', 'test_json.json'), 'r')
for line in rp:
    rec = json.loads(line)  
    print(rec)  
    
    if 'fname' in rec.keys():
        print(rec['fname'])
        
    else:
        print("fname not found")
    
    if 'age' in rec.keys():
        print(rec.get('age'))
        
    else:
        print("age not found")    
        
    print("----------")   
rp.close()    
```

using a function to read

```python
import os
def process(file_path):
    if os.path.exists(file_path):
        rp = open(file_path, 'r')
        for line in rp:
            print(line)
        rp.close()

process('C:\\DSML\\Python_Practise-master\\citi_short_list1.csv') 
```
