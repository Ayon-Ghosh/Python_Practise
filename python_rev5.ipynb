```python
#Return sends a specified value back to its caller whereas Yield can produce a 
#sequence of values. We should use yield when we want to iterate over a sequence, 
#but don't want to store the entire sequence in memory. Yield are used in Python generators.
#---or
```

```python
class gen_even:
    
    
    def __init__(self):
        self.num = ''
        
    def setlimit(self):
        self.num = int(input("enter the limit: "))
        
    def even_num(self):
        i = 0
        while i<self.num:
            if i%2==0:
                yield(i)
            i = i + 1
            
        
```

```python
obj_test = gen_even()
obj_test.setlimit()
x = obj_test.even_num()
for i in x:
    print(i)
```

```python
alist=[2,5,7,9,11,17,222]

def binarySearch(alist, item):
    first_index = 0
    last_index = len(alist) - 1
    found = False
    index = 0
    while first_index<=last_index:
    
        mid_index = (first_index  +  last_index)//2
        if item == alist[mid_index]:
            found = True
            index = mid_index
            return index  
            
        else:
            if item>alist[mid_index]:
                first_index = mid_index + 1
                
            else:
                last_index = mid_index - 1
                
      
    return "Not found"       
        
    
```

```python
import re
password_list = input("enter the passwords separated by comma: ").split(',')
result = []
pattern = re.compile(r'[a-z+A-Z+0-9+$#@+]{6,18}')
for password in password_list:
        if pattern.search(password) is None:
            continue
        else:
            result.append(password)
        
result
```

```python
class CustomerNotAllowed(Exception):

    def __init__(self, value):
        self.value = value

    def __str__(self):
        return repr(self.value)
```
